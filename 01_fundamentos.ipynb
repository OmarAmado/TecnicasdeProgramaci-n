```python
nombre= input("Escribe tu nombre completo:")
edad= input("Escribe tu edad:")
es_mayor= int(edad) >= 18
if es_mayor:
    print(nombre + " tiene " + str(edad) + " años y es mayor de edad")
else:
    print(nombre + " tiene " + str(edad) + " años y es menor de edad")
```
